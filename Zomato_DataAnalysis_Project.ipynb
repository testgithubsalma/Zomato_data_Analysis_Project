```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

# Step 2 - Create the data Frame

```python
dataframe = pd.read_csv("Zomato data .csv")
print(dataframe)
```

```python
dataframe
```

# Convert the data type of column - rate

```python
def handleRate(value):
    value = str(value).split('/')
    value = value[0];
    return float(value) 
dataframe['rate'] = dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

```python
dataframe.info()
```

# Type of restaurant

```python
dataframe.head()
```

```python
colors = ['red', 'green', 'orange', 'blue']
sns.countplot(
    x='listed_in(type)',
    data=dataframe,
    hue='listed_in(type)',
    palette=colors,
    legend=False
)

plt.xlabel("Type of Restaurant")
plt.show()
```

 # Conclusion - majority of the restaurant falls in dinning category

```python
dataframe.head()
```

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})
plt.plot(result, c = "green", marker = "o")
plt.xlabel("Type of restaurant", c = "red", size = 20)
plt.ylabel("Votes", c = "red", size = 20)
```

 # conclusion - dinning restaurants has recieved maximum votes

```python
dataframe.head()
```

```python
plt.hist(dataframe['rate'], bins = 5) 
plt.title("ratings distributions")
plt.show()
```

# Conclusion - the majority restaurants recieved ratings from 3.5 to 4

```python
dataframe.head()
```

```python
sns.set_style("whitegrid")

sns.countplot(
    x='approx_cost(for two people)',
    data=dataframe,
    hue='approx_cost(for two people)',
    palette='husl',   # 🌈 rainbow gradient
    legend=False
)

plt.xticks(rotation=90)
plt.xlabel("Approx Cost (for two people)")
plt.show()
```

# Conclusion - tye majority of couples prefer restaurants with an approximate cost of 300 rupees 

# which mode receives maximum rating

```python
plt.figure(figsize = (6,6))
sns.boxplot(x=  'online_order', y = 'rate', data = dataframe)
```

# conclusion - offline received lower rating in comparison to online order

```python
dataframe.head()
```

```python
pivot_table = dataframe.pivot_table(index= 'listed_in(type)', columns = 'online_order', aggfunc = 'size', fill_value = 0)
sns.heatmap(pivot_table, annot = True, cmap = "YlGnBu", fmt = 'd')
plt.title("Heatmap")
plt.xlabel("Online order")
plt.ylabel("Listed In (Type)")
plt.show()
```

# CONCLUSION: Dining restaurants primarily accept offline orders, whereas cafes primarily receive online orders. This suggests that clients preferred orders in person at restaurants, but prefer online ordering at cafes.
